```python
import xlwings as xw
import json

file_path = r"C:\Users\user\Desktop\산재보험\2차 코호트 1~5차년도 산재보험패널조사 데이터 및 코드북\2차 코호트 5차년도 CODEBOOK_통합 DATA(2022).xlsx"

app = xw.App(visible=False)
wb = app.books.open(file_path)

variable_dict = {}

for sheet in wb.sheets:
    print("\n" + "="*60)
    print(f"📄 시트: {sheet.name}")
    # 💡 A3, B3 출력
    try:
        a3 = sheet.range("A3").value
        b3 = sheet.range("B3").value
        print(f"🔎 A3 (변수명 예시): {a3}")
        print(f"🔎 B3 (변수설명 예시): {b3}")
    except Exception as e:
        print(f"❌ A3/B3 읽기 실패: {e}")
    try:
        col_a = sheet.range("A3:A1000").value
        col_b = sheet.range("B3:B1000").value
    except Exception as e:
        print(f"❌ 시트 불러오기 실패: {e}")
        continue

    if not col_a or not col_b:
        print("⚠️ 열이 비어 있음 → 건너뜀")
        continue

    count = 0
    for i, (var, desc) in enumerate(zip(col_a, col_b), start=3):
        if var and desc:
            clean_var = str(var).replace("**", "").strip()
            clean_desc = str(desc).strip()
            variable_dict[clean_var] = clean_desc
            count += 1
            print(f"  [{i}행] 변수명: {clean_var} / 설명: {clean_desc}")
        else:
            print(var)
            print(desc)
            print(f"  [{i}행] 짝 중 하나가 없음 → 건너뜀")

    print(f"✅ 시트 '{sheet.name}'에서 {count}개 추가")

wb.close()
app.quit()

# 저장
save_path = r"C:\Users\user\Desktop\산재보험\variable_dict.json"
with open(save_path, "w", encoding="utf-8") as f:
    json.dump(variable_dict, f, ensure_ascii=False, indent=4)

print(f"\n🎉 딕셔너리 저장 완료! → {save_path}")
print(f"🧾 전체 변수 수: {len(variable_dict)}")
```

```python
import pandas as pd

# CSV 불러오기
file_path = r"C:\Users\user\Desktop\산재보험\2차\Data.csv"
df = pd.read_csv(file_path, encoding='utf-8-sig')

### -------- 1차 조건 -------- ###

# 조건1: 근로기간 x or 반복 갱신 가능 or 
cond1_part1 = (df['E1023001'] == 2) | (df['E2024001'] == 2)
cond1_part2 = (
    ((df['E1023001'] == 1) | (df['E2024001'] == 1)) &
    ((df['E1026001'] == 3) | (df['E2027001'] == 3))
)
#자영업자 소득 o, 주기적 일자리 o
cond1_part3 = (df['E3014001'] == 1) | (df['E3011007'] == 1)
cond1 = cond1_part1 | cond1_part2 | cond1_part3

# 조건2: 그만둘 의사
cond2 = (df['E1042001'] == 2) | (df['E2041001'] == 2) | (df['E3021001'] == 1)

# 조건3: 급여
cond3 = (df['E1017001'] >= 200) | (df['E2019001'] >= 200)

# first_yo 조건 결과를 원본 df에 Boolean 열로 추가
df['first_yo'] = cond1 & cond2

# 조건 만족 행 수 출력
print(f"1️⃣ 조건 1을 만족하는 행 수: {cond1.sum()}개")
print(f"2️⃣ 조건 2를 만족하는 행 수: {cond2.sum()}개")
print(f"3️⃣ 조건 3을 만족하는 행 수: {cond3.sum()}개")
print(f"✅ 조건1 & 조건2를 만족하는 행 수: {df['first_yo'].sum()}개")
```

```python
# 1차 조건 만족하는 행만 필터링
first_yo_df = df[df['first_yo']]

# 조건1: 근속기간 12개월 이하 또는 결측
cond1_second = (
    (first_yo_df['E1013002'] <= 12) |
    (first_yo_df['E2015002'] <= 12) |
    (first_yo_df['E1013002'].isna() & first_yo_df['E2015002'].isna())
)

# 조건2: 특정 산업/직종
pairs = [
    ('E1019003', 'E2021002'),
    ('E1019007', 'E2021006'),
    ('E1019011', 'E2021010'),
    ('E1019015', 'E2021014'),
    ('E1019017', 'E2021016'),
    ('E1021002', 'E2022001'),
    ('E1021003', 'E2022002'),
    ('E1021004', 'E2022003')
]
cond2_second = pd.Series(False, index=first_yo_df.index)
for col1, col2 in pairs:
    cond2_second |= (first_yo_df[col1] == 1) | (first_yo_df[col2] == 1)

# 최종 조건
second_condition = cond1_second & cond2_second

# 원본 df에 second_yo 열로 추가
df['second_yo'] = False
df.loc[first_yo_df.index, 'second_yo'] = second_condition

# 결과 출력
print(f"✅ 'second_yo' 조건을 만족하는 행 수: {df['second_yo'].sum()}개")
```

```python
# 1. 먼저 삭제할 열 이름 조건에 맞는 것들 제거
drop_cols = []

for col in df.columns:
    if (
        col in ['pid', 'wave', 'nonresponse','E1023001', 'E2024001', 'E1026001', 'E2027001','E3014001', 'E3011007','E1042001', 'E2041001', 'E3021001','E1017001', 'E2019001']
        or col.startswith('p')
        or col.startswith('emp')
        or col.startswith('wt')
    ):
        drop_cols.append(col)

df_cleaned = df.drop(columns=drop_cols, errors='ignore')  # 없는 컬럼 있어도 에러 안나게
#df_cleaned.to_csv(r'C:\Users\user\Desktop\산재보험\2차\df_cleaned.csv', index=False, encoding='utf-8-sig')
```

```python

# 2. E1~E6 데이터프레임 생성
prefixes = ['E1', 'E2', 'E3', 'E4', 'E5', 'E6']
result_dfs = {}
cleaned_dfs = {}

for prefix in prefixes:
    # 해당 prefix로 시작하는 열들만 추출
    cols = [col for col in df_cleaned.columns if col.startswith(prefix)]
    
    # 해당 열들 중 결측이 아닌 값이 2개 이상인 행만 필터링
    filtered_rows = df_cleaned[cols].notna().sum(axis=1) >= 2
    result_dfs[prefix] = df_cleaned.loc[filtered_rows, cols].copy()

# 3. 다른 prefix 제거하고 자기 prefix만 남기기
for prefix in prefixes:
    df_sub = result_dfs[prefix]
    keep_cols = [col for col in df_sub.columns if col.startswith(prefix)]
    cleaned_dfs[prefix] = df_sub[keep_cols].copy()

# 4. 결과 출력: 행 개수
print("📊 각 데이터프레임 행 개수:")
total_rows = 0
for prefix in prefixes:
    row_count = len(cleaned_dfs[prefix])
    print(f"▶ {prefix}: {row_count}개")
    total_rows += row_count

print(f"\n✅ 총합: {total_rows}개")

# 5. 각 데이터프레임 변수로 저장 (선택)
E1, E2, E3, E4, E5, E6 = [cleaned_dfs[p] for p in prefixes]
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

chi2_all_results = {}
target_col = 'first_yo'
alpha = 0.05

# 조건에 사용된 열들
used_columns = [
    'E1023001', 'E2024001', 'E1026001', 'E2027001',
    'E3014001', 'E3011007',
    'E1042001', 'E2041001', 'E3021001',
    'E1017001', 'E2019001','second_yo'
]

# 실제 분석 대상이 되는 데이터프레임들
dataframes = {'E1': E1, 'E2': E2, 'E3': E3}

for name, df_sub in dataframes.items():
    # 조건에 사용된 열 + 타겟 열 제거
    categorical_cols = [col for col in df_sub.columns if col not in used_columns and col != target_col]

    chi2_results = []
    for col in categorical_cols:
        if df_sub[col].isnull().all():
            continue  # 전부 NaN이면 스킵

        contingency = pd.crosstab(df_sub[col], df_sub[target_col])
        if contingency.shape[0] > 1 and contingency.shape[1] > 1:
            try:
                chi2, p, dof, expected = chi2_contingency(contingency)
                if p < alpha:
                    chi2_results.append((col, chi2, p))
            except:
                continue

    chi2_df = pd.DataFrame(chi2_results, columns=['변수명', '카이제곱통계량', 'p값'])
    chi2_df = chi2_df.sort_values(by='카이제곱통계량', ascending=False)
    chi2_all_results[name] = chi2_df

    print(f"\n📊 {name} - p값 < {alpha} 인 유의 변수 TOP 10:")
    print(chi2_df.head(10))
```
