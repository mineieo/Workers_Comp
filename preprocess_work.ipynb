```python
import xlwings as xw
import json

file_path = r"C:\Users\user\Desktop\μ‚°μ¬λ³΄ν—\2μ°¨ μ½”νΈνΈ 1~5μ°¨λ…„λ„ μ‚°μ¬λ³΄ν—ν¨λ„μ΅°μ‚¬ λ°μ΄ν„° λ° μ½”λ“λ¶\2μ°¨ μ½”νΈνΈ 5μ°¨λ…„λ„ CODEBOOK_ν†µν•© DATA(2022).xlsx"

app = xw.App(visible=False)
wb = app.books.open(file_path)

variable_dict = {}

for sheet in wb.sheets:
    print("\n" + "="*60)
    print(f"π“„ μ‹νΈ: {sheet.name}")
    # π’΅ A3, B3 μ¶λ ¥
    try:
        a3 = sheet.range("A3").value
        b3 = sheet.range("B3").value
        print(f"π” A3 (λ³€μλ… μμ‹): {a3}")
        print(f"π” B3 (λ³€μμ„¤λ… μμ‹): {b3}")
    except Exception as e:
        print(f"β A3/B3 μ½κΈ° μ‹¤ν¨: {e}")
    try:
        col_a = sheet.range("A3:A1000").value
        col_b = sheet.range("B3:B1000").value
    except Exception as e:
        print(f"β μ‹νΈ λ¶λ¬μ¤κΈ° μ‹¤ν¨: {e}")
        continue

    if not col_a or not col_b:
        print("β οΈ μ—΄μ΄ λΉ„μ–΄ μμ β†’ κ±΄λ„λ€")
        continue

    count = 0
    for i, (var, desc) in enumerate(zip(col_a, col_b), start=3):
        if var and desc:
            clean_var = str(var).replace("**", "").strip()
            clean_desc = str(desc).strip()
            variable_dict[clean_var] = clean_desc
            count += 1
            print(f"  [{i}ν–‰] λ³€μλ…: {clean_var} / μ„¤λ…: {clean_desc}")
        else:
            print(var)
            print(desc)
            print(f"  [{i}ν–‰] μ§ μ¤‘ ν•λ‚κ°€ μ—†μ β†’ κ±΄λ„λ€")

    print(f"β… μ‹νΈ '{sheet.name}'μ—μ„ {count}κ° μ¶”κ°€")

wb.close()
app.quit()

# μ €μ¥
save_path = r"C:\Users\user\Desktop\μ‚°μ¬λ³΄ν—\variable_dict.json"
with open(save_path, "w", encoding="utf-8") as f:
    json.dump(variable_dict, f, ensure_ascii=False, indent=4)

print(f"\nπ‰ λ”•μ…”λ„λ¦¬ μ €μ¥ μ™„λ£! β†’ {save_path}")
print(f"π§Ύ μ „μ²΄ λ³€μ μ: {len(variable_dict)}")
```

```python
import pandas as pd

# CSV λ¶λ¬μ¤κΈ°
file_path = r"C:\Users\user\Desktop\μ‚°μ¬λ³΄ν—\2μ°¨\Data.csv"
df = pd.read_csv(file_path, encoding='utf-8-sig')

### -------- 1μ°¨ μ΅°κ±΄ -------- ###

# μ΅°κ±΄1: κ·Όλ΅κΈ°κ°„ x or λ°λ³µ κ°±μ‹  κ°€λ¥ or 
cond1_part1 = (df['E1023001'] == 2) | (df['E2024001'] == 2)
cond1_part2 = (
    ((df['E1023001'] == 1) | (df['E2024001'] == 1)) &
    ((df['E1026001'] == 3) | (df['E2027001'] == 3))
)
#μμμ—…μ μ†λ“ o, μ£ΌκΈ°μ  μΌμλ¦¬ o
cond1_part3 = (df['E3014001'] == 1) | (df['E3011007'] == 1)
cond1 = cond1_part1 | cond1_part2 | cond1_part3

# μ΅°κ±΄2: κ·Έλ§λ‘ μμ‚¬
cond2 = (df['E1042001'] == 2) | (df['E2041001'] == 2) | (df['E3021001'] == 1)

# μ΅°κ±΄3: κΈ‰μ—¬
cond3 = (df['E1017001'] >= 200) | (df['E2019001'] >= 200)

# first_yo μ΅°κ±΄ κ²°κ³Όλ¥Ό μ›λ³Έ dfμ— Boolean μ—΄λ΅ μ¶”κ°€
df['first_yo'] = cond1 & cond2

# μ΅°κ±΄ λ§μ΅± ν–‰ μ μ¶λ ¥
print(f"1οΈβƒ£ μ΅°κ±΄ 1μ„ λ§μ΅±ν•λ” ν–‰ μ: {cond1.sum()}κ°")
print(f"2οΈβƒ£ μ΅°κ±΄ 2λ¥Ό λ§μ΅±ν•λ” ν–‰ μ: {cond2.sum()}κ°")
print(f"3οΈβƒ£ μ΅°κ±΄ 3μ„ λ§μ΅±ν•λ” ν–‰ μ: {cond3.sum()}κ°")
print(f"β… μ΅°κ±΄1 & μ΅°κ±΄2λ¥Ό λ§μ΅±ν•λ” ν–‰ μ: {df['first_yo'].sum()}κ°")
```

```python
# 1μ°¨ μ΅°κ±΄ λ§μ΅±ν•λ” ν–‰λ§ ν•„ν„°λ§
first_yo_df = df[df['first_yo']]

# μ΅°κ±΄1: κ·Όμ†κΈ°κ°„ 12κ°μ›” μ΄ν• λλ” κ²°μΈ΅
cond1_second = (
    (first_yo_df['E1013002'] <= 12) |
    (first_yo_df['E2015002'] <= 12) |
    (first_yo_df['E1013002'].isna() & first_yo_df['E2015002'].isna())
)

# μ΅°κ±΄2: νΉμ • μ‚°μ—…/μ§μΆ…
pairs = [
    ('E1019003', 'E2021002'),
    ('E1019007', 'E2021006'),
    ('E1019011', 'E2021010'),
    ('E1019015', 'E2021014'),
    ('E1019017', 'E2021016'),
    ('E1021002', 'E2022001'),
    ('E1021003', 'E2022002'),
    ('E1021004', 'E2022003')
]
cond2_second = pd.Series(False, index=first_yo_df.index)
for col1, col2 in pairs:
    cond2_second |= (first_yo_df[col1] == 1) | (first_yo_df[col2] == 1)

# μµμΆ… μ΅°κ±΄
second_condition = cond1_second & cond2_second

# μ›λ³Έ dfμ— second_yo μ—΄λ΅ μ¶”κ°€
df['second_yo'] = False
df.loc[first_yo_df.index, 'second_yo'] = second_condition

# κ²°κ³Ό μ¶λ ¥
print(f"β… 'second_yo' μ΅°κ±΄μ„ λ§μ΅±ν•λ” ν–‰ μ: {df['second_yo'].sum()}κ°")
```

```python
# 1. λ¨Όμ € μ‚­μ ν•  μ—΄ μ΄λ¦„ μ΅°κ±΄μ— λ§λ” κ²ƒλ“¤ μ κ±°
drop_cols = []

for col in df.columns:
    if (
        col in ['pid', 'wave', 'nonresponse','E1023001', 'E2024001', 'E1026001', 'E2027001','E3014001', 'E3011007','E1042001', 'E2041001', 'E3021001','E1017001', 'E2019001']
        or col.startswith('p')
        or col.startswith('emp')
        or col.startswith('wt')
    ):
        drop_cols.append(col)

df_cleaned = df.drop(columns=drop_cols, errors='ignore')  # μ—†λ” μ»¬λΌ μμ–΄λ„ μ—λ¬ μ•λ‚κ²
#df_cleaned.to_csv(r'C:\Users\user\Desktop\μ‚°μ¬λ³΄ν—\2μ°¨\df_cleaned.csv', index=False, encoding='utf-8-sig')
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

chi2_all_results = {}
target_col = 'first_yo'
alpha = 0.05

# μ΅°κ±΄μ— μ‚¬μ©λ μ—΄λ“¤
used_columns = [
    'E1023001', 'E2024001', 'E1026001', 'E2027001',
    'E3014001', 'E3011007',
    'E1042001', 'E2041001', 'E3021001',
    'E1017001', 'E2019001','second_yo'
]

# μ‹¤μ  λ¶„μ„ λ€μƒμ΄ λλ” λ°μ΄ν„°ν”„λ μ„λ“¤
dataframes = {'E1': E1, 'E2': E2, 'E3': E3}

for name, df_sub in dataframes.items():
    # μ΅°κ±΄μ— μ‚¬μ©λ μ—΄ + νƒ€κ² μ—΄ μ κ±°
    categorical_cols = [col for col in df_sub.columns if col not in used_columns and col != target_col]

    chi2_results = []
    for col in categorical_cols:
        if df_sub[col].isnull().all():
            continue  # μ „λ¶€ NaNμ΄λ©΄ μ¤ν‚µ

        contingency = pd.crosstab(df_sub[col], df_sub[target_col])
        if contingency.shape[0] > 1 and contingency.shape[1] > 1:
            try:
                chi2, p, dof, expected = chi2_contingency(contingency)
                if p < alpha:
                    chi2_results.append((col, chi2, p))
            except:
                continue

    chi2_df = pd.DataFrame(chi2_results, columns=['λ³€μλ…', 'μΉ΄μ΄μ κ³±ν†µκ³„λ‰', 'pκ°’'])
    chi2_df = chi2_df.sort_values(by='μΉ΄μ΄μ κ³±ν†µκ³„λ‰', ascending=False)
    chi2_all_results[name] = chi2_df

    print(f"\nπ“ {name} - pκ°’ < {alpha} μΈ μ μ λ³€μ TOP 10:")
    print(chi2_df.head(10))
```
